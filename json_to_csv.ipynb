```python
import pandas as pd
import numpy as np
import os
import json
```

```python
folder_path = "C:/Users/bhavi/Desktop/women-t20i-cricket-data-analysis/data/raw/t20s_female_json"
```

```python
records = []
```

```python
for file in os.listdir(folder_path):
    if file.endswith(".json"):
        with open(os.path.join(folder_path, file), "r", encoding="utf-8") as f:
            data = json.load(f)

        info = data["info"]

        record = {
            "match_id": file.replace(".json", ""),
            "date": info["dates"][0],
            "team1": info["teams"][0],
            "team2": info["teams"][1],
            "venue": info.get("venue"),
            "city": info.get("city"),
            "winner": info.get("outcome", {}).get("winner"),
            "toss_winner": info.get("toss", {}).get("winner"),
            "toss_decision": info.get("toss", {}).get("decision"),
            "event": info.get("event", {}).get("name")
        }

        records.append(record)

matches_df = pd.DataFrame(records)
matches_df.head()
```

```python
india_df = matches_df[
    (matches_df["team1"] == "India") |
    (matches_df["team2"] == "India")
]

india_df.shape
```

```python
matches_df.to_csv("C:/Users/bhavi/Desktop/women-t20i-cricket-data-analysis/data/processed/match_level.csv", index=False)
```

```python
india_df.to_csv("C:/Users/bhavi/Desktop/women-t20i-cricket-data-analysis/data/processed/india_t20i_matches.csv", index=False)
```
