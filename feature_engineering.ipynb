```python
import pandas as pd

india_df = pd.read_csv("C:/Users/bhavi/Desktop/women-t20i-cricket-data-analysis/data/processed/india_t20i_matches.csv")
india_df.head()
```

```python
india_df["opponent"] = india_df.apply(
    lambda row: row["team2"] if row["team1"] == "India" else row["team1"],
    axis=1
)
```

```python
def result(row):
    if row["winner"] == "India":
        return "Win"
    elif pd.isna(row["winner"]):
        return "No Result"
    else:
        return "Loss"

india_df["india_result"] = india_df.apply(result, axis=1)
```

```python
india_df["date"] = pd.to_datetime(india_df["date"])
india_df["year"] = india_df["date"].dt.year
```

```python
india_df["india_won_toss"] = india_df["toss_winner"] == "India"
```

```python
def india_batted_first(row):
    if row["toss_winner"] == "India" and row["toss_decision"] == "bat":
        return True
    elif row["toss_winner"] != "India" and row["toss_decision"] == "field":
        return True
    else:
        return False

india_df["india_batted_first"] = india_df.apply(india_batted_first, axis=1)
```

```python
india_df.to_csv(
    "data/processed/india_t20i_matches_enriched.csv",
    index=False
)
```
